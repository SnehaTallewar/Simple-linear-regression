```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score , mean_squared_error, mean_absolute_error
```

```python
data = pd.read_csv("Salary_data.csv")
```

```python
plt.scatter(data['Salary'], data['YearsExperience'], color = 'lightblue')
plt.show()
```

```python
 x = data[['YearsExperience']]
 y = data[['Salary']]
```

```python
x_train , x_test , y_train , y_test = train_test_split( x, y , test_size = 0.2 , random_state = 0)
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
```

```python
test = lr_model.predict(x_test)
train = lr_model.predict(x_train)
```

```python
print(test)
```

```python
plt.scatter(x_test, y_test, label="Actual Salary", color="orange")
plt.plot(x_test, test, label="Predicted Salary", color="black", linewidth=2)
```

```python
print(r2_score(y_test , test))
print(mean_absolute_error(y_test , test))
print(mean_squared_error(y_test , test))
```
